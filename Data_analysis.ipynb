```python
import pandas as pd
import time
import lyricsgenius
import json
import os
import csv
from tqdm import tqdm

# =========================
# FILTER + BALANCE SET
# =========================
df_filtered = df_with_genres[df_with_genres['artist_popularity'] > 0]

allowed_countries = [
    'US', 'GB', 'DE', 'CA', 'FR', 'IT', 'AT', 'ES', 'BR', 'SE',
    'NL', 'JP', 'MX', 'JM', 'RU', 'FI', 'PL', 'SK', 'PH', 'PR'
]
df_filtered = df_filtered[df_filtered['country'].isin(allowed_countries)]

df_balanced = (
    df_filtered
    .groupby('country', group_keys=False)
    .apply(lambda x: x.sort_values('artist_popularity', ascending=False).head(3000))
    .reset_index(drop=True)
)

df_balanced = df_balanced[~df_balanced['name'].str.contains(r"K\.|Concerto|Sonata|Op\.|Symphony", case=False, na=False)]


df_balanced['id'] = df_balanced['id'].astype(str)

print(f"Balanced dataset shape: {df_balanced.shape}")
print(df_balanced['country'].value_counts())

# =========================
# LYRICS SCRAPER SETUP
# =========================
ACCESS_TOKEN = "zXW7zwPGmNl49_OeVGejpKAH05FgfSvf3st1YtHpFbwLx6kib_QzCGIhqcQampTv"
GENIUS_REQUEST_DELAY = 1.2
CACHE_FILE = "lyrics_cache.json"
OUTPUT_FILE = "df_with_lyrics.csv"

genius = lyricsgenius.Genius(ACCESS_TOKEN, timeout=10, retries=3)
genius.skip_non_songs = True
genius.excluded_terms = ["(Remix)", "(Live)"]
genius.remove_section_headers = True

# Load cache
if os.path.exists(CACHE_FILE):
    with open(CACHE_FILE, "r") as f:
        lyrics_cache = json.load(f)
else:
    lyrics_cache = {}

# Load processed IDs
processed_ids = set()
if os.path.exists(OUTPUT_FILE):
    with open(OUTPUT_FILE, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row.get('lyrics') and len(row['lyrics']) > 50:
                processed_ids.add(row['id'])

print(f" Already processed songs: {len(processed_ids)}")


df = df_balanced[~df_balanced['id'].isin(processed_ids)].copy()
print(f" Remaining songs to process: {df.shape[0]}")

# =========================
# MAIN LOOP
# =========================
header_written = os.path.exists(OUTPUT_FILE)

for idx, row in tqdm(df.iterrows(), total=df.shape[0], desc="Scraping lyrics"):
    track_id = row['id']
    title = row['name']
    artist = row['artists']

    if isinstance(artist, list):
        artist = ", ".join(artist)

    if track_id in lyrics_cache:
        lyrics = lyrics_cache[track_id]
    else:
        lyrics = None
        for attempt in range(3):
            try:
                song = genius.search_song(title, artist)
                if song:
                    lyrics = song.lyrics
                break
            except Exception as e:
                print(f" Retry {attempt+1}/3 for {artist} - {title}")
                if hasattr(e, "response") and e.response:
                    if e.response.status_code == 429:
                        wait_time = int(e.response.headers.get("Retry-After", 60))
                        print(f"🔁 Rate limit – waiting {wait_time} sec...")
                        time.sleep(wait_time)
                    else:
                        print(f" HTTP {e.response.status_code}: {e}")
                        break
                else:
                    print(f" Error: {e}")
                    time.sleep(30)
        lyrics_cache[track_id] = lyrics
        with open(CACHE_FILE, "w") as f:
            json.dump(lyrics_cache, f)
        time.sleep(GENIUS_REQUEST_DELAY)


    row_data = row.to_dict()
    row_data['lyrics'] = lyrics

    with open(OUTPUT_FILE, "a", encoding="utf-8", newline='') as f:
        writer = csv.DictWriter(f, fieldnames=row_data.keys())
        if not header_written:
            writer.writeheader()
            header_written = True
        writer.writerow(row_data)

print(" Lyrics scraping completed")
```
